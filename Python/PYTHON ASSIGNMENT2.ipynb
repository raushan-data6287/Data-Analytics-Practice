```python
# QUES-13: CREATE A CODE TO COUNT THE OCCURENCES OF EACH ELEMENT IN A LIST AND RETURN A DICTIONARY WITH ELEMENTS AS KEYS AND THEIR COUNT AS VALUES.

my_list = [1, 2, 3, 4,"a", "b", True, False,True]

count_dict = {}

for item in my_list:
    count = my_list.count(item)
    count_dict[item] = count

print(count_dict)
```

```python
# QUES-14:  WRITE A CODE TO REVERSE A LIST IN-PLACE WITHOUT USING ANY BUILT-IN REVERSE FUNCTION.
my_list = [1, 2, 3, 4, "raushan", "rahul", 5+6j, True, False]

start = 0
end = len(my_list) - 1

while start < end:
    my_list[start] , my_list[end] = my_list[end] , my_list[start]
    start += 1
    end -= 1

print(my_list)
```

```python
# QUES-16: CREATE A CODE TO CHECK IF GIVEN LIST IS SORTED(EITHER IN ASCENDING OR DESCENDING ORDER) OR NOT.
user_input = input("ENTER YOUR NUMBER: ")

my_list = list(map(int, user_input.split()))

if my_list == sorted(my_list):
    print("LIST IS SORTED IN ASCENDING ORDER")
elif my_list == sorted(my_list, reverse = True):
    print("LIST IS SORTED IN DESCENDING ORDER")
else:
    print("LIST IS NOT SORTED")


print("YOUR LIST", my_list)
```

```python
# QUES-18: IMPLEMENT A CODE TO FIND THE INTERSECTION OF TWO GIVEN LISTS.

list1 = [1,2,4,5,7,3]
list2 = [2,5,6,7,3,9]

intersection = [] # same values in both list1 and list2.

for num in list1:
    if num in list2:
        intersection.append(num)

print(intersection)
```

```python
# QUES-20: WRITE A CODE TO SUFFLE A GIVEN LIST RANDOMLY WITHOUT USING ANY BUILT-IN SHUFFLE FUNCTION.

import random

my_list = [1,2,3,4,5,6,7,8]

n = len(my_list)

for i in range(n-1, 0, -1):   
    j = random.randint(0, i) 
    my_list[i], my_list[j] = my_list[j], my_list[i]  

print("Shuffled List:", my_list)
```
