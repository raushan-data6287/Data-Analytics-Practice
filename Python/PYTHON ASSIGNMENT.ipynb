```python
# QUES-2: WRITE A CODE TO COUNT THE NUMBER OF VOVEL IN A STRING.

name = "raushan kumar"
vowels = "aeiouAEIOU"

COUNT = 0
for ch in name:
    if ch in vowels:
        COUNT += 1 

print("total vowels in name is ", COUNT)
```

```python
# QUES-3: WRITE A CODE TO CHECK IF A GIVEN STRING IS A PALINDROME OR NOT.
text = input("enter your string value")
reverse_text = text[::-1]

if text == reverse_text:
    print("It is a palindrome")
else:
    print("It is not a palindrome")   
```

```python
# QUES-4: WRITE A CODE TO CHECK IF TWO GIVEN STRINGS ARE ANAGRAMS OF EACH OTHER.
text1 = input("enter your first string")
text2 = input("enter your second string")

if sorted(text1) == sorted(text2):
    print("It is a anagrams")
else:
    print("It is not a anagrams")
```

```python
# QUES-5: WRITE A CODE TO FIND ALL OCCURENCES OF A GIVEN SUBSTRING WITHIN ANOTHER STRING.
main_string = input("enter your main string")
sub_string = input("enter your substring")

indexes = []
m = len(main_string)
n = len(sub_string)

for i in range(m - n+1):
    if main_string[i : i+n] == sub_string:
        indexes.append(i)

print("occurance", indexes)
print("total", len(indexes))

```

```python
# QUES-6: WRITE A CODE TO PERFORM BASIC STRING COMPRESSION USING THE COUNTS OF REPEATED CHARACTER.
S = input("Enter your string: ")
result = ""
count = 1

for i in range(len(S) - 1):
    if S[i] == S[i+1]:
        count += 1
    else:
        result += S[i] + str(count)
        count = 1


result += S[-1] + str(count)

print("Compressed:", result)

```
